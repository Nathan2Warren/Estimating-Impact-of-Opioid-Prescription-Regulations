```python
import pandas as pd
import numpy as np
csa = pd.read_csv('https://raw.githubusercontent.com/nickeubank/practicaldatascience/master/Example_Data/ca/ca_arrests_2009.csv')
pop = pd.read_csv('https://raw.githubusercontent.com/nickeubank/practicaldatascience/master/Example_Data/ca/nhgis_county_populations.csv')
```

```python
csa.head()
```

```python
pop.head()
```

```python
popcali = pop[pop.STATE == 'California']
pop2009 = popcali[popcali.YEAR == '2005-2009']
pop2009.head()
```

```python
csa_correct = csa.copy()
csa_correct.loc[csa.COUNTY == 'Del Norte County','COUNTY'] = 'DelNorte County'
csa_correct.loc[csa.COUNTY == 'Inyo County','COUNTY'] = 'Injo County'                
dataset2009 = pd.merge(csa_correct, pop2009, how= 'left',on='COUNTY', validate='1:1',indicator=True)
dataset2009
```

```python
dataset2009['violent_arrest_rate_2009'] = (dataset2009.VIOLENT * 100 / dataset2009.total_population)
dataset2009['drug_offenses_rate_2009'] = (dataset2009.F_DRUGOFF * 100 / dataset2009.total_population)
```

```python
dataset2009.head()
```

```python
### import 2018 Arrests Data 
arrest2018 = pd.read_csv('https://raw.githubusercontent.com/nickeubank/practicaldatascience/master/Example_Data/ca/ca_arrests_2018.csv')
arrest2018.head()
```

```python
### get population data from 2013-2017
pop1 = pop[pop.STATE == 'California']
pop2018 = pop1[pop1.YEAR == '2013-2017']
pop2018.head()
```

```python
### Merging 2018 Dataset
arrest2018_correct = arrest2018.copy()
arrest2018_correct.loc[arrest2018.COUNTY == 'Del Norte County','COUNTY'] = 'DelNorte County'
arrest2018_correct.loc[arrest2018.COUNTY == 'Inyo County','COUNTY'] = 'Injo County'                
dataset2018 = pd.merge(arrest2018_correct, pop2018, how= 'left',on='COUNTY', validate='1:1',indicator=True)
dataset2018
```

```python
dataset2018['violent_arrest_rate_2018'] = (dataset2018.VIOLENT * 100 / dataset2018.total_population)
dataset2018['drug_offenses_rate_2018'] = (dataset2018.F_DRUGOFF * 100 / dataset2018.total_population)
dataset2018.head(10)
```

```python
### Exercise 14
print(dataset2018['drug_offenses_rate_2018'].mean())
print(dataset2009['drug_offenses_rate_2009'].mean())

## As you can see, the drug arrest goes down from 2009 to 2018
```

```python
### Exercise 15
print(dataset2009['violent_arrest_rate_2009'].mean())
print(dataset2018['violent_arrest_rate_2018'].mean())

print(dataset2009['violent_arrest_rate_2009'].mean() - dataset2018['violent_arrest_rate_2018'].mean())

## Violent crime did decrease following drug legalization and the average violent arrest rate decrease 0.0283% from 2009 t0 2018.
```

```python
### Split sample into two groups
high_group = dataset2009[dataset2009.drug_offenses_rate_2009 > dataset2009['drug_offenses_rate_2009'].mean()]
low_group = dataset2009[dataset2009.drug_offenses_rate_2009 <= dataset2009['drug_offenses_rate_2009'].mean()]
```

```python
rateavg2009 = dataset2009['violent_arrest_rate_2009'].mean()
highrateavg = high_group['violent_arrest_rate_2009'].mean()
lowrateavg = low_group['violent_arrest_rate_2009'].mean()
```

```python
### Exercise 16
high_diff = highrateavg - rateavg2009
low_diff = lowrateavg - rateavg2009
diffindiff = high_diff - low_diff
print(diffindiff)

## Violent crime did fall more from 2009 to 2018 in the counties that had lots of drug arrests in 2009 (where legalization likely had more of an effect) than in counties with fewer drug arrests in 2009 (where legalization likely mattered less).
```

```python
### Exercise 17
high_perchange = high_diff / rateavg2009
low_perchange = low_diff / rateavg2009
diffperchange = high_perchange - low_perchange
print(diffperchange)

## There also a greater proportionate decline
```
